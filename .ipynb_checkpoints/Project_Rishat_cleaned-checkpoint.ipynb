# A Hedonic Analysis of Housing Prices in Moscow

## Research question
How do apartment characteristics such as size, metro accessibility, region, apartment type, and renovation affect housing prices in Moscow?

## Hypothesis
Larger apartments, better renovation, and shorter distance to metro are associated with higher prices.


## 1. Data and project structure

This project studies apartment-level listing data for Moscow and the Moscow region. The analysis follows a simple hedonic-pricing approach:

1. load and prepare the dataset using reusable functions from `src/housing_utils.py`;
2. visualise key relationships in the data;
3. estimate a baseline OLS model for log housing prices;
4. interpret the main findings and discuss limitations.

This notebook is intentionally kept as the **narrative layer** of the project, while the reusable logic lives in `src/housing_utils.py`.


```python
from src.housing_utils import (
    load_data,
    prepare_features,
    plot_area_vs_price,
    plot_price_by_renovation,
    fit_baseline_model,
)

df = load_data("data/data_moscow.csv")
df = prepare_features(df)

print(df.shape)
df.head()
```

```python
df[[
    "price", "area", "minutes_to_metro", "number_of_rooms",
    "living_area", "kitchen_area", "floor", "number_of_floors",
    "price_per_sqm", "floor_ratio"
]].describe().round(2)
```

## 2. Feature engineering

The main transformed variables used in the analysis are:

- `log_price`: the natural logarithm of apartment price;
- `log_area`: the natural logarithm of total apartment area;
- `price_per_sqm`: price divided by apartment area;
- `floor_ratio`: apartment floor divided by the total number of floors in the building.

Using logarithms makes the relationship between apartment size and price easier to model and interpret.


## 3. Visualisation

```python
plot_area_vs_price(df)
```

The scatter plot shows a strong positive relationship between apartment area and price in log terms. This suggests that size is likely to be one of the most important predictors of housing prices in the baseline model.


```python
plot_price_by_renovation(df)
```

The boxplot indicates that renovation status is associated with meaningful differences in price per square meter. More upgraded apartments tend to be priced higher, although the categories may also capture differences in apartment segment and location.


## 4. Regression model

The baseline specification is:

\[
\log(\text{price}) = f(\log(\text{area}),\ \text{minutes to metro},\ \text{region},\ \text{apartment type},\ \text{renovation},\ \text{floor ratio})
\]

This is a cross-sectional hedonic model, so the objective is to identify associations between housing attributes and prices rather than causal effects.


```python
model = fit_baseline_model(df)
model.summary()
```

## 5. Interpretation of results

The model explains a large share of the variation in prices, with an \(R^2\) of about **0.91**, which suggests that the selected housing characteristics are highly informative for predicting listing prices.

The coefficient on `log_area` is strongly positive and statistically significant, confirming that apartment size is the most important predictor in the model. Because both variables are in logs, this coefficient can be interpreted approximately as an elasticity: larger apartments are associated with disproportionately higher prices.

`minutes_to_metro` has a negative and statistically significant coefficient, indicating that apartments located farther from the metro tend to be cheaper.

The coefficient for `Moscow region` is negative relative to the reference category `Moscow`, which suggests that apartments outside Moscow proper are associated with lower prices, all else equal.

`Apartment type` and `renovation` also matter a lot. In particular, better renovation categories are associated with higher prices. One interesting result is that `Without renovation` is still more expensive than the baseline renovation category in this model, which may reflect market segmentation: some higher-end apartments may be sold unfinished and then customised by buyers.

Finally, `floor_ratio` is positive and significant, but its effect is much smaller than the effects of size, location, and renovation.


## 6. Limitations and next steps

This project has several important limitations.

First, the dataset is **cross-sectional**, so it does not measure true housing **price dynamics** over time.

Second, the prices are likely **listing prices**, not necessarily final transaction prices, which means the dependent variable may not fully reflect realised market outcomes.

Third, location is only partly observed through `region` and `minutes_to_metro`. A richer study would include exact coordinates, neighbourhood fixed effects, distance to the city centre, and local amenities.

Fourth, renovation and apartment type may partly capture omitted quality differences, so the coefficients should not be interpreted as causal effects.

If I were to continue this project, I would look for transaction-level data with dates, richer geospatial variables, and potentially estimate alternative specifications such as price per square meter models or neighbourhood-level fixed effects.


## 7. Conclusion

Overall, the results show that housing prices in Moscow are strongly associated with apartment size, metro accessibility, region, and renovation status. The project supports a clear hedonic-pricing interpretation: both physical housing characteristics and accessibility-related attributes matter for listing prices.
